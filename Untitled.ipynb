```sos
proc sql;
create table collapsed as
select
  count(*) as count,
  date,
  katben
from long
group by katben, date;
quit;
```

```sos
data collapsed_utf8 (encoding=utf-8);
set collapsed;
run;

proc print data=collapsed_utf8 (obs=10);
run;
```
